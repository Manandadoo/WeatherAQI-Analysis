```python
import pandas as pd

# Function to calculate AQI for a single pollutant based on its concentration and breakpoints
def calculate_individual_aqi(concentration, breakpoints):
    for (C_lo, C_hi, I_lo, I_hi) in breakpoints:
        if C_lo <= concentration <= C_hi:
            return ((I_hi - I_lo) / (C_hi - C_lo)) * (concentration - C_lo) + I_lo
    return None

# Breakpoints for PM2.5 (in µg/m³)
pm25_breakpoints = [
    (0, 30, 0, 50), (31, 60, 51, 100), (61, 90, 101, 200),
    (91, 120, 201, 300), (121, 250, 301, 400), (251, 500, 401, 500)
]

# Breakpoints for PM10 (in µg/m³)
pm10_breakpoints = [
    (0, 50, 0, 50), (51, 100, 51, 100), (101, 250, 101, 200),
    (251, 350, 201, 300), (351, 430, 301, 400), (431, 600, 401, 500)
]

# Breakpoints for O3 (in ppb)
o3_breakpoints = [
    (0, 50, 0, 50), (51, 100, 51, 100), (101, 168, 101, 200),
    (169, 208, 201, 300), (209, 748, 301, 400)
]

# Breakpoints for NO2 (in ppb)
no2_breakpoints = [
    (0, 40, 0, 50), (41, 80, 51, 100), (81, 180, 101, 200),
    (181, 280, 201, 300), (281, 400, 301, 400), (401, 1000, 401, 500)
]

# Breakpoints for SO2 (in ppb)
so2_breakpoints = [
    (0, 40, 0, 50), (41, 80, 51, 100), (81, 380, 101, 200),
    (381, 800, 201, 300), (801, 1600, 301, 400), (1601, 10000, 401, 500)
]

# Breakpoints for CO (in mg/m³)
co_breakpoints = [
    (0, 1, 0, 50), (1.1, 2, 51, 100), (2.1, 10, 101, 200),
    (10.1, 17, 201, 300), (17.1, 34, 301, 400), (34.1, 50, 401, 500)
]

# Function to calculate the overall AQI for a row based on pollutant concentrations
def calculate_aqi_for_row(row):
    # Calculate AQI for each pollutant using the dataset's specific field names
    pm25_aqi = calculate_individual_aqi(row['PM2.5 (ug/m3)'], pm25_breakpoints) if 'PM2.5 (ug/m3)' in row else None
    pm10_aqi = calculate_individual_aqi(row['PM10 (ug/m3)'], pm10_breakpoints) if 'PM10 (ug/m3)' in row else None
    o3_aqi = calculate_individual_aqi(row['Ozone (ug/m3)'], o3_breakpoints) if 'Ozone (ug/m3)' in row else None
    no2_aqi = calculate_individual_aqi(row['NO2 (ug/m3)'], no2_breakpoints) if 'NO2 (ug/m3)' in row else None
    so2_aqi = calculate_individual_aqi(row['SO2 (ug/m3)'], so2_breakpoints) if 'SO2 (ug/m3)' in row else None
    co_aqi = calculate_individual_aqi(row['CO (mg/m3)'], co_breakpoints) if 'CO (mg/m3)' in row else None

    # Return a dictionary of AQI values for each pollutant
    return {
        'PM2.5 AQI': pm25_aqi,
        'PM10 AQI': pm10_aqi,
        'Ozone AQI': o3_aqi,
        'NO2 AQI': no2_aqi,
        'SO2 AQI': so2_aqi,
        'CO AQI': co_aqi,
        'Overall AQI': max(filter(None, [pm25_aqi, pm10_aqi, o3_aqi, no2_aqi, so2_aqi, co_aqi]), default=None)
    }

# Assuming df is your dataframe with hourly AQI data
# Apply the function to calculate AQI for each row and store results in a new dataframe
aqi_results = numeric_dataset.apply(calculate_aqi_for_row, axis=1, result_type='expand')

# Combine the original dataframe with the AQI results
numeric_dataset = pd.concat([numeric_dataset, aqi_results], axis=1)

# Print the AQI values for each parameter
print(numeric_dataset[['PM2.5 AQI', 'PM10 AQI', 'Ozone AQI', 'NO2 AQI', 'SO2 AQI', 'CO AQI', 'Overall AQI']])
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt

# Feature engineering: add lagged AQI values as features
numeric_dataset['AQI_lag1'] = numeric_dataset['Overall AQI'].shift(1)
numeric_dataset['AQI_lag2'] = numeric_dataset['Overall AQI'].shift(2)

# Drop rows with missing values (due to lag features)
# data = data.dropna()

# Features and target
features = ['tempC', 'humidity', 'windspeedKmph', 'pressure', 'AQI_lag1', 'AQI_lag2']
X = numeric_dataset[features]
y = numeric_dataset['AQI']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Train Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')

# Plot actual vs predicted AQI
# plt.plot(y_test.values, label='Actual AQI')
# plt.plot(y_pred, label='Predicted AQI')
# plt.legend()
# plt.show()

plt.figure(figsize=(14, 7))  # Increase dimensions here (width, height)
plt.plot(numeric_dataset.index[-len(y_test):], y_test.values, label='Actual AQI', color='red')  # Use datetime index for x-axis
plt.plot(numeric_dataset.index[-len(y_pred):], y_pred, label='Predicted AQI', color='blue')  # Use datetime index for x-axis
plt.title('Actual vs Predicted AQI', fontsize=16)  # Adding title
plt.xlabel('Time', fontsize=14)  # Label for x-axis
plt.ylabel('AQI', fontsize=14)  # Label for y-axis
plt.legend()
plt.grid()
plt.show()
```
